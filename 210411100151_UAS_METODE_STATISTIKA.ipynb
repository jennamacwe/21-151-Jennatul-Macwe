# INTERVAL KEPERCAYAAN




> interval kepercayaan adalah metode menghitung dan memperkirakan keakuratan sebuah sampel yang mewakili data yang sesungguhnya yang dapat diperkirakan dengan Point Estimate (titik acuan yang digunakan untuk memperkirakan suatu nilai parameter interval, dapat berupa rata rata dan proporsi) dan Interval Estimate (selang dari nilai sebenarnya mungkin berbeda)

> Interval kepercayaan berguna untuk mengukur variant sampel dalam populasi dan dapat memprediksikan letak nilai yang sebenarnya dari populasi tersebut

> Interval kepercayaan terdiri dari
tiga bagian yaitu nilai statistik (rerata atau proporsi), tingkat kepercayaan, dan derajat penyimpangan (standar error)

**cara menghitung interval kepercayaan**

Screenshot 2022-06-28 125911.jpg

Keterangan :

* CI = Interval Kepercayaan
* P = proporsi sampel
* Z = tingkat kepercayaan
* N = jumlah sampel

Konstanta tingkat kepercayaan (confidence level)
* confidence level 99%, Z = ± 2.575
* confidence level 95%, Z = ± 1.96
* confidence level 90%, Z = ± 1.645
* confidence level 80%, Z = ± 1.28

```python
import statistics
populasi=[50,58,43,64,47,50,56,47,60,64]
print('Data = ', populasi)

rata = statistics.mean(populasi)
print(f"Mean / rata-rata = ({' + '.join(f'{x}' for x in populasi)}) / {len(populasi)}")
print(f'Mean / rata-rata = {rata}')
```

# STANDAR DEVIASI


> Standar deviasi merupakan suatu nilai statistik yang digunakan untuk menentukan persebaran data dalam sampel, dan seberapa dekat elemen data-data yang ada dengan mean dari sampel tersebut.

> standar deviasi berfungsi untuk menentukan semakin besar selisihnya dengan mean, semakin heterogen data yang terdapat dalam populasi/sampel.

> Variansi standar deviasi dibagi menjadi 2 yaitu (variansi populasi (σ2) dan variansi sampel (S2)), sehingga hal tersebut pun berlaku pada standar deviasi yang merupakan akar dari variansi yaitu standar deviasi populasi (σ) dan standar deviasi sampel (S).

**cara menghitung standar deviasi**

* mencari mean dari data
* menghitung selisih per data dengan mean
* memasukan data ke rumus

**rumus**

> mean = jumlah data / banyak data

# SKEWNESS DAN KURTOSIS

> **skewness** bisa disebut dengan kemiringan merupakan ukuran   yang menyatakan derajat ketidak simetrisan kurva distribusi frekuensi, atau dengan kata lain menunjukkan seberapa jauh   distribusi   itu   menyimpang   dari   simetris   atau   normal.

> menurut segi kemiringannya, skewness dibagi menjadi 3 yaitu:
* Jika koefisien kemiringannya lebih kecil dari nol (< 0), model distribusinya negatif
* Jika koefisien kemiringannya sama dengan nol (= 0), model distribusinya simetris
* jika koefisien kemiringannya lebih besar dari nol (> 0), model distribusinya positif.

**rumus**

> **kurtosis** bisa disebut dengan keruncingan merupakan derajat kelancipan dari distribusi tersebutterhadap distribusi normal (kurva normal)

> menurut segi kelancipannya, kurtosis dibagi menjadi 3 yaitu:
* Jika suatu distribusi (kurva) lebih lancip ataulebih ramping dibandingkan terhadap kurva normal,distribusinya disebut leptokurtis.
* Jika suatu distribusi (kurva) normal, distribusinya disebut mesokurtis 
* Jika suatu distribusi (kurva) lebih landai atau lebih tumpul dibandingkan terhadap kurva normal,distribusinya disebut platikurtis

> **rumus**

Screenshot 2022-06-28 133741.jpg

```python
from scipy.stats import kurtosis, skew
populasi=[50,58,43,64,47,50,56,47,60,64]
print('Data = ',populasi)

skewness = skew(populasi)
kurtosis = kurtosis(populasi)
print(f'Skewness = {skewness}')
print(f'Kurtosis = {kurtosis}')
```
